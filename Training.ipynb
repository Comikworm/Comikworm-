You are given a list of dictionaries representing books in a library. Each
dictionary contains information about a book, including its title,
author, genre, and publication year. Implement a function to perform
linear search to find books published in a specific year. If books are
found, return a list containing titles of books published in that year. If
no books are found, return an empty list.

```python
def find_books_by_year(library, year):
    books_published_in_year = []
    for book in library:
        if book['publication_year'] == year:
            books_published_in_year.append(book['title'])
    return books_published_in_year
```

You are given a list of lists where each inner list represents a row in a
matrix. Implement a function to perform linear search to find the
position of a specific element in the matrix. If the element is found,
return its row and column indices. If the element is not found, return
(-1, -1).

```python
def find_element_in_matrix(matrix, element):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == element:
                return (i, j)
    return (-1, -1)
```

```python
def linear_search_matrix(matrix, target):
    for i in range(len(matrix)):          # row index
        for j in range(len(matrix[i])):   # column index
            if matrix[i][j] == target:
                return (i, j)
    return (-1, -1)
```

You are given a sorted list of integers and a target value. Implement a
function to perform binary search to find the index of the closest
element to the target value in the list. If there are multiple closest
elements, return the index of the smallest one.

```python
def binary_search_closest(arr, target):
    left, right = 0, len(arr) - 1
    closest_index = -1
    min_difference = float('inf')

    while left <= right:
        mid = (left + right) // 2
        difference = abs(arr[mid] - target)

        if difference < min_difference:
            min_difference = difference
            closest_index = mid

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return closest_index
```

```python
def binary_search_closest(arr, target):
    low = 0
    high = len(arr) - 1

    # Edge cases
    if target <= arr[0]:
        return 0
    if target >= arr[-1]:
        return len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    # After loop, high < low
    # Compare arr[high] and arr[low]
    if abs(arr[high] - target) <= abs(arr[low] - target):
        return high
    else:
        return low
```

You are given a sorted list of integers and a target value. Implement a
function to perform binary search to find the index of the last
occurrence of the target value in the list. If the target value is found,
return its index. If the target value is not found, return -1.

```python
def binary_search_hard(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            left = mid + 1  # Continue searching on the right side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return result
```

```python
def binary_search_last_occurrence(arr, target):
    low = 0
    high = len(arr) - 1
    result = -1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            result = mid        # store index
            low = mid + 1       # search right side for last occurrence
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return result
```

You are given a sorted list of integers and a target value. Implement a
function to perform binary search to find the index of the target value
in the list. If the target value is found, return its index. If the target
value is not found, return the index where it would be inserted in order
to maintain the sorted order.

```python
def binary_search_insertion_index(arr, target):
    left, right = 0, len(arr)
    while left < right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid
    return left
```

```python
def binary_search_insert_position(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    # If not found, low is the insert position
    return low
```
